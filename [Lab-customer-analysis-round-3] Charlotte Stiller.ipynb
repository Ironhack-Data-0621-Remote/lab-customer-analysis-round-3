## 1. Show DataFrame info.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

#importing the dataframe + showing the info

data = pd.read_csv('marketing_customer_analysis.csv')
data.info
```

## 2. Describe DataFrame.


```python
#describe dataframe with the describe function
data.describe().T
```

```python
#standrdizing header names with the lower function + replacing the space with an underscore
data.columns = data.columns.str.lower()
data.columns = data.columns.str.replace(' ','_')
data
```

## 3. How a plot of the total number of responses.


```python
#counting the number of response
data["response"].value_counts()
```

```python
#puting the responses in da dataframe
count_response = pd.DataFrame((data["response"]).value_counts())
count_response
```

```python
#creating countplot for responses

plt.figure(figsize=(4,4))

sns.countplot("response", data=data)
plt.legend(title="Reponse marketing customer analysis")
plt.show()
```

## 4. Show a plot of the response rate by the sales channel.


```python
#checking witch column I have to use

data.columns
```

```python
#creating barplot for responses by sales channels

plt.figure(figsize=(10,4))
actual = sns.countplot('sales_channel', hue='response', data=data)

# giving the tilte for the plot
plt.title("Show a plot of the response rate by the sales channel")

# naming the c and y axis 
plt.xlabel("sales chanel")
plt.ylabel("response")

# creating the legend of the bars in the plot
plt.legend(title="Reponse by sales channels")

# saving the plot as a 'png'
plt.savefig('ResponeSalesChanelPlot.png')

#showing the plot
plt.show()
```

## 5. Show a plot of the response rate by the total claim amount.


```python
#creating barplot for responses by the total claim amount. 
#Woundering what the error "No handles with labels found to put in legend" means, because it seems to work

plt.figure(figsize=(6,4))

sns.barplot(x="total_claim_amount", y="response", data=data)

plt.legend(title="Reponse by total claim amount")
plt.show()
```

## 6. Show a plot of the response rate by income.

```python
#creating barplot for responses by income
plt.figure(figsize=(4,2))

sns.barplot(x="response", y="income", data=data)
plt.legend(title="Reponse by income")
plt.show()
```
